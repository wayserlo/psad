#Задача 4.3

Правительство города М. испытывает систему обнаружения нежелаемых лиц по камерам в метро. В качестве демонстрации работоспособности системы была поставлена цель: найти и задержать опасного рецидивиста по имени Николай Вальный, а также его соучастников. Была собрана выборка из 5000 снимков лица, для которых была проверена гипотеза о несовпадении этого снимка с лицами участников команды Н. Вального. Для 100 фотографий нулевая гипотеза была отвергнута на уровне значимости alpha = 0.05.

Требуется:

1. Определить, в чем недостаток данного подхода и как можно его улучшить.
2. Предложить наилучший, на ваш взгляд, способ для повышения качества данного решения.
3. Какую меру качества контролирует данный способ? Какие гарантии он предоставляет?
3. В чём недостатки данного способа?
4. Как изменилась мощность при использовании предложенного вами способа относительно изначальной процедуры проверки?
5. Известно, что все 5000 фотографий были сделаны для разных людей, и правительство хочет, чтобы система ни в коем случае не упустила преступников. Ответьте на те же вопросы из пунктов 2-4.
6. Как изменилась мощность при использовании предложенного вами способа относительно предыдущего способа?

Все выкладки должны быть сделаны аналитически, без использования компьютера. (допускается использование компьютера для подстановки численных значений в финальную формулу)



---

#Решение

1. В данном случае у нас задача множественной проверки гипотез. А именно: у нас $m=5000$ выборок, для каждой из которых мы проверяем нулевую гипотезу (о несовпадении лица с лицами команды Н. Вального) на уровне значимости $\alpha = 0.05$. Т.е. вероятность ошибочно отклонить верную $H_0$ при проверке только одной фотографии равна $0.05$, а если мы будем проверять $5000$ фотографий, то вероятность, что среди этого множества ошибочно отклонить $H_0$ уже становится равным $m_0 * \alpha$  (где $m_0$ - число фотографий, для которых $H_0$ верно), т.е. если мы не подправим уровни значимости для каждой из гипотез, то вполне вероятно, что неверно распознаем человека на фотографии за человека из команды Н. Вального. 

  Для контроля вводится *групповая вероятность ошибки* (многомерное обобщение ошибки первого рода). Обозначим за $V$ число отвергнутых $H_{0i}$ при условии, что $H_{0i}$ были верными. Тогда групповая вероятность ошибки $FWER = P(V >0)$. Мы должны сохранить эту ошибку на уровне $\alpha$, т.е. $FWER = P(V >0) \leq \alpha$. Делается это за счет корректировки уровней значимости $\alpha_i$ для каждой из гипотез $i = 1, \dots m$. 

  На лекции было предложено много методов для корректировки $\alpha_i$, например: **поправка Бонферрони, метод Холма (нисходящие метод), метод Шидака и его модификация метод Шикада-Холма (нисходящие метод) и др..** Если не делать дополнительных предположений и характере зависимостей статистик, то можно улучшить результаты с помощью метода Холма
2. Я считаю, что для данной задачи для правительства города М. гораздо важнее задержать Н. Вального и они готовы будут дополнительно проверить людей, которых ошибно приняли за людей из команды Н. Вального. Я веду к тому, что методы Бонферрони и Холма слишком редко начнут отвергать $H_0$ из-за очень маленьких $\alpha_i$, следовательно, критерии станут очень маломощными. Для контроля этого аспекта лучше пользоваться восходящими методами множественной проверки гипотез. Без дополнительных предположений о характере зависимости статистик, считаю лучшим методом для данной задачи - метод Бенджамини-Иекутиели:
$$ \alpha_i = \dfrac{i\cdot\alpha}{m\sum\limits_{j=1}^{m}\frac{1}{j}}$$
3. Метод обеспечивает контроль над **ожидаемой долей ложных отклонений гипотез FDR** на уровне значимости $\frac{m_0}{m}\alpha$:
$$ FDR = \mathbb{E}\left(\dfrac{V}{\max(R,1)}\right) \leq \dfrac{m_0}{m}\alpha$$
При отсутствии информации о зависимости между статистиками метод неулучшаем.
4. Данный метод дает более плохие результаты для мощности по сравнению с методом Бенджамини-Хохберга. 
5. Мощность уменьшится. Мощность $W =\dfrac{S}{m-m_0}$, где $S$ - число верно отвергнутых нулевых гипотез. При уменьшении $\alpha_i$ мы станем реже отклонять гипотезы, значит, с большей вероятностью не отклоним $H_0$ и в тех случаях, когда это требуется. $S$ уменьшится (или не изменится в идеальном варианте развития событий), следовательно, $W$ уменьшится
6. В данном случае статистики $T_i$ независимы. *Только я изначально начала считать, что система не хочет упускать преступников, так что, может быть, мои ответы на вопросы выше не будут совпадать с ожидаемым*

    2.   Используем метод Бенджамини-Хохберга
          $$ \alpha_i = \dfrac{i * \alpha}{m}$$
    3.   Так же контролирует $FDR$, но уже на уровне $\alpha$. Более мощный по сравнению с Бенджамини-Иекутиели (и со всеми нисходящими методами и поправкой Бонферрони)
    4.  $FWER$ будет выше, чем у Бенджамини-Иекутиели, требует условия независимости статистик 
7. Мощность увеличилась, у данного метода лучшие показатели мощности относительно всех предыдущих рассмотренных методов.
